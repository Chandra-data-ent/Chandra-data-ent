```python
#@title **Update and view the previous Matches** { display-mode: "form" }
import pandas as pd
import seaborn as sns
from datetime import date
import matplotlib.pyplot as plt
from google.colab import files
from google.colab import files
file_path = "/matches.csv"
matches = pd.read_csv(file_path)

matches
```

```python
#@title **Enter the match innings id and see the winner!!** { display-mode: "form" }
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

match_id = input("Enter the Match ID: ")
match_info = matches[matches["id"] == int(match_id)]

if match_info.empty:
    print("No information available for the given Match ID.")
else:
    team1 = match_info["team1"].values[0]
    team2 = match_info["team2"].values[0]
    winner = match_info["winner"].values[0]
    date = match_info["date"].values[0]
    venue = match_info["venue"].values[0]

    print("Match ID:", match_id)
    print("Date on which the match happened:", date)
    print("Venue of the match:", venue)
    print("Team 1:", team1)
    #print("Score of Team 1:", score)
    print("Team 2:", team2)
    print("Winner:", winner)
```

```python
#@title **Don't know the match id of the innings ? Dont worry!! Enter the YEAR of the IPL Season to view all the matches !!** { display-mode: "form" }

year = input("Enter the year of the IPL season: ")
season_matches = matches[matches["Season"].str.contains(year)]
if season_matches.empty:
    print("No matches found for the given year.")
else:
    print("Matches played in the IPL", year, "season:")
    for index, match in season_matches.iterrows():
        print("Match ID:", match["id"])
        print("Team 1:", match["team1"])
        print("Team 2:", match["team2"])
        print("---------------------------")
```

```python
#@title **Enter the season/year and see which teams played and who won the title!!** { display-mode: "form" }

import pandas as pd

year = input("Enter the year of the IPL season: ")

teams_playing = matches[matches["Season"].str.contains(year)]["team1"].unique()

if len(teams_playing) == 0:
    print("No teams found for the given year.")
else:
    print("Teams playing in the IPL", year, "season:")
    for i, team in enumerate(teams_playing, start=1):
        print(f"{i}. {team}")

    last_match = matches[matches["Season"].str.contains(year)].tail(1)
    if not last_match.empty:
        last_match_winner = last_match["winner"].values[0]
        print("Winner of the last match in the IPL", year, "season:", last_match_winner)
```
